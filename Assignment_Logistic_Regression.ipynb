Question 1 :  What is Simple Linear Regression (SLR)? Explain its purpose.

Answers : Simple Linear Regression (SLR) is a statistical method used to model and analyze the relationship between two variables â€” one independent variable (X) and one dependent variable (Y) â€” by fitting a straight line (called the regression line) through the data points.

Simple Linear Regression estimates how much the dependent variable (Y) changes when the independent variable (X) changes by one unit.

The relationship is expressed using the regression equation:

$$
Y = \beta_0 + \beta_1 X + \epsilon
$$




Questions 2 :  What are the key assumptions of Simple Linear Regression?

Answer : Key Assumptions of Simple Linear Regression

1. Linearity

The relationship between the independent variable (X) and the dependent variable (Y) is linear.
That means, when you plot Y vs. X, the points roughly follow a straight line.

$$
Y = \beta_0 + \beta_1 X + \epsilon
$$

2. Independence of Errors

The residuals (errors) are independent of each other.
This means one observationâ€™s error should not depend on anotherâ€™s.

3. Homoscedasticity (Constant Variance of Errors)

The variance of residuals is constant across all levels of X.
In other words, the spread of the residuals should be roughly the same everywhere.

4. Normality of Errors

The residuals (errors) are normally distributed.
This is especially important for making statistical inferences (like confidence intervals and hypothesis tests).

5. No (or minimal) Measurement Error in X

Questions 3 :  Write the mathematical equation for a simple linear regression model and
explain each term.

Answer :

$y = \beta_0 + \beta_1 x + \varepsilon$

> Explanation of Each Term
- y: The dependent variable (or response variable) â€” the value we are trying to predict or explain.
- x: The independent variable (or predictor variable) â€” the value used to make predictions.
- \beta _0: The intercept â€” the expected value of y when x=0.
- \beta _1: The slope coefficient â€” it represents the change in y for a one-unit increase in x.
- \varepsilon : The error term â€” captures the difference between the observed value and the value predicted by the model (i.e., the residuals).

Question 9: Write Python code to fit a simple linear regression model using scikit-learn

Answer :

```python
#Answer :

from sklearn.linear_model import LinearRegression
import numpy as np

# Sample data: X = independent variable, y = dependent variable
X = np.array([[1000], [1500], [2000], [2500], [3000]])
y = np.array([200000, 250000, 300000, 350000, 400000])

# Create and fit the model
model = LinearRegression()
model.fit(X, y)

# Print slope and intercept
print("Slope (Î²1):", model.coef_[0])
print("Intercept (Î²0):", model.intercept_)
```

Question 10: How do you interpret the coefficients in a simple linear regression model?

Answer:  In a simple linear regression model, the relationship between a dependent variable y and an independent variable x is modeled as:

   # y = b0 + b1*x + e

 Coefficient Interpretations
- Intercept (\beta _0):
- This is the predicted value of y when x=0.
- It provides a baseline level of the dependent variable.
- Example: If \beta _0=50, then when x=0, the model predicts y=50.
- Slope (\beta _1):
- This represents the change in y for a one-unit increase in x.
- It tells you the direction and magnitude of the relationship.
- If \beta _1>0: y increases as x increases.
- If \beta _1<0: y decreases as x increases.
- Example: If \beta _1=3, then for every 1-unit increase in x, y increases by 3 units.
